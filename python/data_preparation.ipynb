```python
import pandas as pd
```

```python
file_path = "CS_data_digi_channels.xlsx" # load data
df_data = pd.read_excel(file_path, sheet_name = "Data")
df_ranks = pd.read_excel(file_path, sheet_name = "Ranks")
```

```python
# check data
df_data.info()
df_ranks.info()
```

```python
# join tables data & ranks
df = df_data.merge(
    df_ranks,
    on = "Country",
    how = "left"
)
```

```python
# check data after merge/join
df.info()
df["Rank asc"].isna().mean()
```

```python
df["Entity_Type"] = df["Age of Athlete"].apply(
    lambda x: "Athlete" if pd.notna(x) else "Team"
)
```

```python
# encoding/decoding
df["Name of Athlete or Team"] = (
    df["Name of Athlete or Team"]
    .astype(str)
    .str.encode("latin1", errors = "ignore")
    .str.decode("utf-8", errors = "ignore")
)
```

```python
# creating age groups
df["Age_Group"] = pd.cut(
    df["Age of Athlete"],
    bins = [0, 20, 25, 30, 35, 40, 100],
    labels= ["<20", "20–24", "25–29", "30–34", "35–39", "40+"]
)
```

```python
# medal score - purpose core analytics in BI
medal_score = {
    "gold": 3,
    "silver": 2,
    "bronze": 1
}
df["Medal Score"] = df["Medal"].map(medal_score)
```

```python
# creating groups "decade" for BI visualisation
df["Decade"] = (df["Year"] // 10) * 10
```

```python
# replace " " by "_"
df.columns = df.columns.str.replace(" ", "_")
```

```python
df.to_csv("Olympics_gold_dataset.csv", index= False)
df.to_excel("Olympics_gold_dataset.xlsx", index=False)
```
